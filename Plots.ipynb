```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sharp v2

```python
x = [15,15,15,15,15,15,15,15,17,19,21,23,25,23,26,28,30,32,34]

def f(x):
    return (((((x-5)/(abs(x-5)+1))-1)/2)*(-x)) 

y = list(map(f, x))

plt.plot(x,markevery=[12],marker='o',markersize=12)
plt.ylim(10,32)
plt.xlim(10,16)
plt.xticks([])
plt.yticks([])
plt.savefig("v2.jpg")
```

# Sharp v1+v2

```python
x = np.linspace(-6, 6, 100)
def f(x):
    return (((abs(x-5.2)/ (x-5.2)) -1) *2*np.sin(x)) + (((( abs(x-5.2) / (x-5.2)) +1) /2) * (np.exp(x/5) +0.705))

y = list(map(f, x))

plt.plot(x, y,markevery=[88],marker='o',markersize=12)
plt.xlim(4,6)
plt.ylim(2,5)
plt.xticks([])
plt.yticks([])
plt.savefig("v1v2.jpg")
```

# Sharps

```python
x = np.linspace(-5, 5, 100)

def f(x):
    return (.2)*(x-1)*(2*x-2)*(2*x-1.3)*(x-4)*(-1)+3

y = list(map(f, x))

plt.plot(x,y,markevery=[57],marker='o',markersize=12)
plt.xlim(0,2)
plt.ylim(0,5)
plt.xticks([])
plt.yticks([])
plt.savefig("v123.jpg")
```

# Sharps + Dent + Smooth

```python
x=np.linspace(-10,10,100)

def f(x):
    return -1*(x-1)*(x-2)+5

y = list(map(f, x))
plt.plot(x,y,markevery=[57],marker='o',markersize=12)
plt.xticks([])
plt.yticks([])
plt.savefig("sv123_dent_smooth.jpg")
```

# Sharps + Dent

```python
x = np.linspace(-5, 5, 100)

def f(x):
    return (.2)*(x-1)*(x+2)*(x-2)*(x-4)*(-1)+3

y = list(map(f, x))

plt.plot(x,y,markevery=[39],marker='o',markersize=12)
plt.xlim(-3,5)
plt.ylim(0,10)
plt.xticks([])
plt.yticks([])
plt.savefig("sharpv123_dent.jpg")
```
