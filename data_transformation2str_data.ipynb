```python
from SPEP import spep
import os
import numpy as np
```

```python
def my_iterator(number):
    for i in range(number):
        yield i
```

```python
directory = 'datasets'
```

```python
# Selecting mask and reference images
ref_name = 'datasets/dataset_1/A/A-kappa/Z_80080808_0230_IFIX1.jpg'
albumin_mask_name = 'datasets/dataset_1/A/A-kappa/Z_80090997_0230_IFIX1.jpg'

ref = spep(ref_name)
mask = spep(albumin_mask_name)

ref.read()
mask.read()

spep.albumin_reference(mask) # Finding albumin mask
spep.barLenght_reference(ref) # Finding height of the bars
spep.bar_finder(ref) # Finding distances of the bars
```

```python
input_size_w_ud = 1778
input_size_wo_ud = 1486
graph_size = 184

inputs_w_ud_w_sp = np.empty([input_size_w_ud, graph_size*6])
inputs_w_ud_w_sp_f = np.empty([input_size_w_ud, graph_size*6])
inputs_w_ud_wo_sp = np.empty([input_size_w_ud, graph_size*5])
inputs_w_ud_wo_sp_f = np.empty([input_size_w_ud, graph_size*5])
labels_w_ud = np.empty([input_size_w_ud,1])

inputs_wo_ud_w_sp = np.empty([input_size_wo_ud, graph_size*6])
inputs_wo_ud_w_sp_f = np.empty([input_size_wo_ud, graph_size*6])
inputs_wo_ud_wo_sp = np.empty([input_size_wo_ud, graph_size*5])
inputs_wo_ud_wo_sp_f = np.empty([input_size_wo_ud, graph_size*5])
labels_wo_ud = np.empty([input_size_wo_ud,1])

classes = {
    "wo":0,
    "Ak":1,
    "Al":2,
    "Mk":3,
    "Ml":4,
    "Gk":5,
    "Gl":6,
    "ud":7
}
```

```python

# With ud and with SP 
data_range = my_iterator(input_size_w_ud)

for root, dir, files in os.walk(directory):
    if root.endswith("A-kappa") or root.endswith("Ak"):
        label = "Ak"
    if root.endswith("A-lambda") or root.endswith("Al"):
        label = "Al"
    if root.endswith("M-kappa") or root.endswith("Mk"):
        label = "Mk"
    if root.endswith("M-lambda") or root.endswith("Ml"):
        label = "Ml"
    if root.endswith("G-kappa") or root.endswith("Gk"):
        label = "Gk"
    if root.endswith("G-lambda") or root.endswith("Gl"):
        label = "Gl"
    if root.endswith("without"):
        label = "wo"
    if root.endswith("unclear decision"):
        label = "ud"

    for file in files:
        file_dir = os.path.join(root,file)

        if file_dir.endswith('.jpg'):
            index = next(data_range)
            obj = spep(file_dir)
            obj.read()
            obj.evaluate()
            obj.find_graphs()
            values = np.empty(0)
            values_f = np.empty(0)
            for key,value in obj.graphs.items():
                np_value = np.array(value)
                np_value_f = np.fft.fft(np_value)
                values = np.concatenate((values, np_value),axis=None)
                values_f = np.concatenate((values_f, np_value_f),axis=None)
            inputs_w_ud_w_sp[index] = values
            inputs_w_ud_w_sp_f[index] = values_f
            labels_w_ud[index] = classes[label]
            
```
